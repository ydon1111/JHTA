```python
x1_data = [73.,93.,89.,96.,73.]
x2_data = [80.,88.,91.,98.,66.]
x3_data = [75.,93.,90.,100.,70.]
y_data = [152.,185.,180.,196.,142.]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
x = tf.placeholder(tf.float32)
#y = w1*x1 + w2*x2 + w3*x3 + b 
#  
w1 = tf.Variable(tf.random_normal([1],name='weight1'))
w2 = tf.Variable(tf.random_normal([1],name='weight2'))
w3 = tf.Variable(tf.random_normal([1],name='weight3'))
b =  tf.Variable(tf.random_normal([1],name='bias'))

model = w1*x1 + w2*x2 + w3*x3 + b 
print(model)
```

```python
model = tf.matmul(X,W) + b 
loss = tf.reduce_mean(tf.square(model-Y))

optimizer = tf.train.GradientDescentOptimizer(learning_rate=1e-5)
train = optimizer.minimize(loss)
# 
sess = tf.Session()
sess.run(tf.global_variables_initializer())


for step in range(2001):
    loss_val ,model_val, _ =sess.run([loss,model,train],
                                 feed_dict={X:x_data,Y:y_data})
    if step%10 ==0:
        print('loss : ', loss_val)

                
print(sess.run([W,b]))
```

```python
tf.set_random_seed(999)
```

```python
x_data = [[2],[4],[6],[8],[10]]
#0  1  
```

```python
y_data = [[0],[0],[0],[1],[1]]
```

```python
x_data ,y_data
```

```python
X = tf.placeholder(tf.float32, shape=[None,1])
Y = tf.placeholder(tf.float32, shape=[None,1])
W = tf.Variable(tf.random_normal([1]), name='weight')
b = tf.Variable(tf.random_normal([1]), name='bias')

```

```python
model = tf.sigmoid(W*X+b)
#sigmoid     
```

```python
loss = - tf.reduce_mean(Y*tf.log(model)+(1-Y)*tf.log(1-model))
train =  tf.train.GradientDescentOptimizer(learning_rate = 0.01).minimize(loss)
```

```python
# sess = tf.Session()
```

```python
#0~1
#0.6
predicted = tf.cast(model>0.6,dtype=tf.float32)
#
accuracy = tf.reduce_mean(tf.cast(tf.equal(predicted,Y),dtype=tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) #  
    for step in range(10001):
        loss_val , _ = sess.run([loss,train],feed_dict={X:x_data,Y:y_data})
        if step %200 == 0:
            print(step, 'loss : ', loss_val)
            
    h , c , a = sess.run([model,predicted,accuracy],feed_dict={X:x_data,Y:y_data})

    print(h,c,a)
    print('-------------------------')
    print('model ',h , '', c , '' ,a)
    print(' ', sess.run([model,predicted],
                           feed_dict={X:[[7],[4],[5],[10],[20]]}))
```

```python
X = tf.placeholder('float',[None,1])
Y = tf.placeholder('float',[None,5])
classCnt = 5 

W = tf.Variable(tf.random_normal([1,classCnt]),name='weight')
b = tf.Variable(tf.random_normal([1,classCnt]),name='bias')

model = tf.nn.softmax(tf.matmul(X,W)+b )
print(model)

## 0 ~1   , tf.nn.softmax 
```

```python
loss = tf.reduce_mean(-tf.reduce_sum(Y*tf.log(model),axis=1))

train = tf.train.GradientDescentOptimizer(learning_rate=0.1).minimize(loss)
```

```python
sess = tf.Session()
sess.run(tf.global_variables_initializer())

for step in range(2001):
    sess.run(train,feed_dict={X:x_data,Y:y_data})
    if step % 200 ==0:
        print(step, sess.run(loss,feed_dict={X:x_data,Y:y_data}))
a = sess.run(model, feed_dict={X:[[3]]})
print(a,sess.run(tf.argmax(a,1)))
```
