```python
# x를 0 부터 20 사이의 정수 
x = np.arange(21)
x
```

```python
p = 1/3 
y = comb(n,x)*p**x*(1-p)**(n-x)  #  이항분포식,확률밀도함수 
y
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline 
plt.rc('font', family = 'Malgun Gothic')
```

```python
plt.figure(figsize=(12,8))
plt.bar(x,y)
plt.xlabel('성공횟수 ')
plt.ylabel('p(x) :  확률 ')
plt.title('이항분포')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y,'o--')
plt.xlabel('성공횟수 ')
plt.ylabel('p(x) :  확률 ')
plt.title('이항분포')
plt.grid()
plt.show()
```

```python
from scipy.stats import binom 
```

```python
mean, var , skew , kurt = binom.stats(n,p,moments ='mvsk')
```

```python
mean, var , skew , kurt
```

```python
y1 = binom(n,p).pmf(x)  #Probability Mass Function) 확률밀도함수,이항분포 
y1
```

```python
plt.figure(figsize=(12,8))
plt.plot(x,y1,'o--')
plt.xlabel('성공횟수 ')
plt.ylabel('p(x) :  확률 ')
plt.title('이항분포')
plt.grid()
plt.show()
```

```python
# 포아송분포
```

$$
     f(x) = \frac{\lambda^x \times e^{-\lambda}}{x!}
$$

```python
from numpy import arange , exp , power 
from scipy.special import factorial 
```

```python
x = arange(41, dtype = 'float64') #큰 숫자를 취급하기 위해 실수형으로 선언 
x
```

```python
l = 20
y = power(l,x) * exp(-1)/factorial(x)
y
```

```python
plt.rc('font', family = 'Malgun Gothic')
```

$$
{\displaystyle F(x;\lambda )=(1-e^{-\lambda x})H(x)}
$$

f(x) = \begin{cases}\lambda e^{-\lambda x} & x > 0\\ 0 & x \leq 0\end{cases}

지수분포의 누적 분포 함수는
$$
{\displaystyle F(x;\lambda )={\begin{cases}1-e^{-\lambda x}&{\mbox{where }}x\geq 0\\0&{\mbox{where }}x<0\end{cases}}}{\displaystyle F(x;\lambda )={\begin{cases}1-e^{-\lambda x}&{\mbox{where }}x\geq 0\\0&{\mbox{where }}x<0\end{cases}}}
$$

```python
import math 
import random 
from collections import Counter
from matplotlib import pyplot as plt 
```

```python
def bernoulli_trial(p):
    return 1 if random.random() < p else 0
```

```python
def binomial(n,p):
    return sum(bernoulli_trial(p) for _ in range(n))
```

```python
def normal_cdf(x,mu=0,sigma=1):
    return (1+ math.erf((x-mu)/math.sqrt(2)/sigma))/2
```

```python
def make_hist(p, n , num_points):
    data = [binomial(n, p) for _ in range(num_points)]
    
    #이항분포의 표본을 막대 그래프로 표현 
    histogram = Counter(data)
    plt.bar([x - 0.4 for x in histogram.keys()], 
            [v / num_points for v in histogram.values()],
            0.8,
            color="0.75")
    
    mu = p*n
    sigma = math.sqrt( n*p*(1-p))
    
    
    # 정규분포를 라인차트로 표현 
    xs = range(min(data), max(data)+1)
    ys = [normal_cdf(i + 0.5, mu, sigma) - normal_cdf(i - 0.5, mu, sigma) for i in xs]
    plt.plot(xs,ys)
    plt.title("정규분포와 이항분포 차이 ")
    plt.show()
    
p=0.5
n= 100
num_points=20 
make_hist(p, n, num_points)
```
