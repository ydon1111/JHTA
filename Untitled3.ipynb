```python
from konlpy.tag import Okt
from collections import Counter

okt = Okt()
nouns = okt.nouns(text)
# print('before ' , nouns)
for i , v in enumerate(nouns):
    if len(v)<2:
        nouns.pop(i)   
# print('after ', nouns)
count = Counter(nouns)
    
# 명사 빈도 카운트
noun_list = count.most_common(100)
for v in noun_list:
    print(v)
```

```python
import csv

with open('noun_list.csv','w',newline = '', encoding='euc-kr') as f:
    csvw = csv.writer(f)
    for v in noun_list:
        csvw.writerow(v)
```

```python
#noun_list.csv 파일 읽어오기
wc = WordCloud(max_font_size = 200 ,
                      background_color='white',
                      height = 1000,
                      width = 1000,
                      font_path = 'c:\\windows\\Fonts\\malgun.ttf')

f = open('noun_list.csv','r',encoding='euc-kr')
news =f.read()
wc.generate(news)
plt.imshow(wc.generate(news) , interpolation='nearest')
plt.axis('off')
```
