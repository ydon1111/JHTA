```python
#정규분포 확률밀도함수
```

$$
    f(x) = \frac{1}{\sqrt{2\pi}\sigma} e^ {- \frac {(x-\mu)^2}{2 \sigma^2}},
    (- \infty < x < \infty)
$$

정규분포
$$
 N( \mu , \sigma^2)
$$


```python
y = (1/ np.sqrt(2 * np.pi))* np.exp( -x**2 /2)
y
```

# t의 pdf 자유도 v 


$$
f(t) = \frac { \Gamma ( \frac {v+1}{2})}{ \Gamma ( \frac {v}{2})}
	 (1+\frac{t^2}{v}) ^ { - \frac {v+1}{2}}  , (-\infty < t < \infty)
$$

```python
import math
```

```python
nu = 10 
y = (math.gamma((nu+1)/2)/ math.gamma(nu/2)*math.sqrt(math.pi*nu))* (1+t**2 /nu) **(-(nu+1)/2)
y
```

```python
import matplotlib.pyplot as plt 
%matplotlib inline 
```

```python
plt.figure(figsize =(10,6))
plt.plot(t,y)
plt.xlabel('y')
plt.grid()
plt.title('t 분포')
plt.legend(['t(10)'])
plt.show()
```

```python
import scipy.stats as stats 
```

```python
y1 = stats.t(10).pdf(t)
y1
```

```python
plt.figure(figsize =(10,6))
plt.plot(t,y1)
plt.xlabel('y')
plt.grid()
plt.title('t 분포')
plt.legend(['t(10)'])
plt.show()
```

```python
mList = []
for i in range(20):
    mList.append(np.random.randint(33,38)+round(np.random.rand(),2))
mList
```

```python
data = pd.Series(mList)
data
```

```python
x_bar = data.mean()
x_bar
```

```python
s = data.std()
s
```

```python
mu = 35
n = 20
```

```python
#귀무가설 : 빵의 평균은 35와 차이가 없다 
#연구가설 (대립가설): 있다 

sample_t = (x_bar - mu) / (s/np.sqrt(float(n)))
sample_t

# 임계값 :
alpha = 0.05 

alpha_t = stats.t.ppf(alpha, n-1)
alpha_t
```

```python
p_value = stats.t.sf(np.abs(sample_t),n-1)
p_value
```

```python
import pandas as pd 
```

```python
np.random.seed(999)
```

```python
# 정규성 검사 보통 그룹당 표본이 30 개 이하의 경우에는 정규성 여부 확인필요
# 극단치가 있는 경우 데이터를 신뢰할 수 없게 되기 때문이다.
# 대한민국 남성의 키를 검정, 3m 이상이 2개 이상 
```

```python
# 등분산은 비교해야 할 대상의 데이터의 분포가 서로 유사한지 알아보기 위해서 
# 데이터 분포가 서로 다르면 통계의 의미가 달라지기 때문이다.
a = []
b = []
[a.append(round(np.random.uniform(170,183),1))for i in range(10)]
[b.append(round(np.random.uniform(175,189),1))for i in range(10)]
a ,b
```

```python
df = pd.DataFrame({'groupA':a, 'groupB':b})
df
```

```python
#정규성 검정 
```

```python
#shpiro 
#귀무가설 : 정규분포를 따른다 
#대립가설 : 정규분포를 따르지 않는다.
from scipy.stats import shapiro 
shapiro(df.groupA)
# 귀무가설을 기각하지 못하므로 정규분포를 따른다 할 수 있다 

#  통계량              p_values
# (0.9312883019447327, 0.460683137178421)
```

```python
shapiro(df.groupB)
```

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
sns.distplot(df.groupA)
plt.title('A group Histogram')
plt.show()
```

```python
sns.distplot(df.groupB)
plt.title('A group Histogram')
plt.show()
```

```python
# 3.등분산 
# 귀무가설 : 두 그룹은 등분산을 만족한다.
# 대립가설 : 두 그룹은 등분산을 만족하지 않는다 .
from scipy.stats import bartlett
bartlett(df.groupA , df.groupB)
# pvalue=0.5276632825411698 > 0.05 보다 크므로 귀무가설 채택 
#등분산이라고 할 수 있다.
```

```python
sns.distplot(df.groupA)
sns.distplot(df.groupB)
plt.show()
```

```python
# 차이를 검정 해보자 .
# 귀무가설 : 두 집단간의 평균차가 없다. 
# 대립가설 : 두 집단간의 평균차가 있다.
statistics , p_value = stats.ttest_ind(
    a=df.groupA,
    b=df.groupB,
    equal_var= True
)

statistics , p_value
#  0.015422940015897349 < 0.05 귀무가설 기각 
```
