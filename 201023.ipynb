```python
#45개 데이터 따로 분리 (행을 출력 )
from sklearn.model_selection import train_test_split
#                                   7:3 비율로 나눔 
X_train , X_test, y_train , y_test = train_test_split(iris.data,iris.target,
                                                     test_size = 0.3,
                                                     random_state=999)
X_train , X_test, y_train , y_test 
```

```python
# 단순회귀모형 
from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
import os 

model = tree.DecisionTreeClassifier()
model.fit(X_train , y_train)
# tree.plot_tree(model)
dot_data = tree.export_graphviz(model,filled=True ,rounded = True , 
                                class_names=iris.target_names,
                               feature_names = iris.feature_names,
                               out_file=None)
graph = graph_from_dot_data(dot_data)
graph
graph.write_png('tree.png')

```

```python
dt_graph = pydotplus.graph_from_dot_data(dot_data)
Image(dt_graph.create_png())
```

```python
df = pd.read_csv('RidingMowers.csv')
# df
```

```python
df.shape 
```

```python
df.keys()
```

```python
x = df.drop(['Ownership'],axis=1)
y = df['Ownership']
```

```python
#문자형 자료를 숫자(범주형)로 인코딩
#숫자형으로 표준화

```

```python
# target 레이블 인코딩 
from sklearn.preprocessing import LabelEncoder  #범주형 변수 수치화
from sklearn.preprocessing import StandardScaler #연속 변수의 표준화

le =  LabelEncoder()
```

```python
y = le.fit_transform(y)

#알파벳순으로 결정 

```

```python
#훈련/검증용 데이터 분할
# 7 : 3 

from sklearn.model_selection import train_test_split
#                                   7:3 비율로 나눔 
# stratify = y  : y의 비율에 따라 분할 

X_train , X_test, y_train , y_test = train_test_split(x,y,
                                                     test_size = 0.3,
                                                     random_state=1,
                                                     stratify=y)
# X_train , X_test, y_train , y_test 
```

```python
#모델 구축
#depth  = 2 

from pydotplus import graph_from_dot_data
from sklearn.tree import export_graphviz
from IPython.display import Image
import pydotplus
import os 

model = tree.DecisionTreeClassifier(max_depth=2 ,random_state=1)
model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_test ,y_pred
```

```python
print('정확도 %.3f'% accuracy_score(y_test,y_pred))
```

```python
confmat = pd.DataFrame(confusion_matrix(y_test,y_pred),index=[['True[0]','True[1]']],
                          columns = ['Predict[0]','Predict[1]'])
confmat
```
